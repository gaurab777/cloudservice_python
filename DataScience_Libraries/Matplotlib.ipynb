# Matplotlib

## 1) Linechart

```python
from matplotlib import pyplot as plt #OR import matplotlib.pyplot as plt
x= [1,2,3,4]
y= [12,20,9,40]
plt.plot(x,y)
plt.show()
```

```python
#Adding labels on x-axis and y-axis, and title to the chart
x = [1,2,3,4]
y = [20,50,33,45]
plt.plot(x,y,color='violet')
plt.xlabel("Serial Number")
plt.ylabel("Distribution")
plt.title("My line chart")
plt.show()
```

## 2) Scatter Plot

```python
x = [18,18,8,14]
y = [12,7,3,9]
plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Class')
plt.title('My scatter plot')
plt.show()
```

```python
# Scatter plot with 1000 random dots
import numpy as np

x = np.random.random(1000)
y = np.linspace(1,100,1000)

plt.scatter(x,y,color='black')
plt.show()
```

## Bar Chart (Bar Plot)

```python
subject = ['Math','Physics','Biology','Chemistry']
marks = [90,80,88,78]
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Bar Plot of a result')
plt.bar(subject,marks,color='green')
plt.show()
```

```python
subjects = ['Math','Physics','Biology','Chemistry']
marks = [90,80,88,78]
plt.xlabel('Marks')
plt.ylabel('Subjects')
plt.title("Horizontal Bar")
plt.barh(subjects,marks)
plt.show()
```

### Comparing two bar charts

```python
plt.bar([2,3,5,7,9],[5,2,3,6,8], label="Example 1")
plt.bar([1,4,6,8,10],[6,5,8,9,6], label="Example 2", color="green")
plt.legend()
plt.xlabel("Number")
plt.ylabel("Height")
plt.title("Comparing bar graph")
plt.show()
```

## Histogram

```python
import numpy as np
a = np.array([22,13,44,4,55,66,77,98,26,44,34,50,99,67,53,67,78])
plt.hist(a, bins = [0,25,50,75,100])
plt.show()
```

```python
x = np.random.randint(1,100,100)
print(x)
plt.hist(x, bins=10)
plt.title('Histogram')
plt.show()
```

## Area Plot/Chart

```python
x= [1,2,3,4]
y= [10,12,4,17]
plt.fill_between(x,y)
plt.xlabel('series')
plt.ylabel('Age')
plt.title('Area chart')
plt.show()
```

## Pie Chart

```python
cars = ['BMW','Merc','Audi','Volkswagen','Porsche']
data = [35,28,22,33,25]
plt.pie(data, labels = cars)
plt.title('German Car Brands')
plt.show()
```

## Boxplot

## Contour Plots

```python
import numpy as np
import matplotlib.pyplot as plt

xlist = np.linspace(-3.0,3.0,100) #Generating 100 numbers
ylist = np.linspace(-3.0,3.0,100) #Generating 100 numbers
X, Y = np.meshgrid(xlist,ylist)#Creating meshgrid
Z = np.sqrt(X**2 + Y**2)
cp = plt.contourf(X,Y,Z)
plt.colorbar(cp) #Adding a colorbar to a plot
plt.show()
```
