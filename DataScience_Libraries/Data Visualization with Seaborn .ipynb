# Data Visualization with Seaborn

 ### Distplots

```python
import matplotlib.pyplot as plt
import seaborn as sb
sb.distplot([0,1,2,3,4,5]) #Distplot function
plt.show()
```

```python
import numpy as np
data = np.random.rand(200) #Creating an array with random values
sb.distplot(data) 
plt.show()
```

### Line Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
x = [10,4,5,13,46,23]
y = [100,220,130,450,500,350]
data_plot = pd.DataFrame({"Product_no":x, "Cost":y})
sb.lineplot(x="Product_no", y="Cost", data=data_plot)
plt.show()
```

### Lmplot

```python
df = sb.load_dataset("anscombe")
print(df)
sb.lmplot(x="x", y="y",data=df)
plt.show()
```

```python
x=[10,20,21,23,25]
y=[50,75,67,89,85]
df = pd.DataFrame({"Roll_no":x, "Marks":y})
print(df)
sb.lmplot(x='Roll_no',y='Marks',data=df)
plt.show()
```

### Marginal Plot

```python
df = sb.load_dataset('iris')
print(df)
#custom the inside plot: options are: "scatter"|"reg"|"resid"|
sb.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind="scatter")
```

### stripplot and swarmplot

```python
#stripplot
tips = sb.load_dataset('tips')
print(tips.head())
sb.stripplot(x="day", y="total_bill", data=tips)
```

```python
#swarmplot
tips = sb.load_dataset('tips')
sb.swarmplot(x="day",y="total_bill",data=tips)
```

### boxplot and violinplot

```python
#Boxplot
sb.boxplot(x="day", y="total_bill", data= tips, palette="rainbow")
```

```python
#violinplot
sb.violinplot(x="day",y="total_bill",data=tips,palette="rainbow")
```

```python
data = np.random.normal(size=(20,6)) + np.arange(6) / 2
sb.set_style("whitegrid") #can also use 'darkgrid'
sb.boxplot(data = data)
plt.title("whitegrid")
plt.show()
```
